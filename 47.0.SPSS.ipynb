SPSS  è sistema di prorammi di calcolo  e statistica -> by IBM

(vecchio nome) Statistical Package for Social Science -> (nuovo nome)Statistical Product and Service Solutions

Esso ha varie procedure come:

*Comandi di definizione, manipolazione e trasformazione dei dati :

Dove, come leggere ed etichettare i dati


*Comandi operativi->applicazione di un numero elevato di procedure statistiche 

Si usano per avere informazioni sul funzionamento di alcuni parametri del programma -> esempio modificare il formato dell'output

*Comandi di procedure statistiche e/o grafiche

I comandi SPSS sono fomarti da una keyword e specifiche opzioni dell'utente

SPSS fa uso di file particolari:

*file dati ->  .sva -> dati che devono essere elaborati

*file sintassi -> .sps -> memorizzano i comandi

*file attivo -> file temporaneo che registra la sessione come variabili, strutture dati

*file output -> .spo -> contiene i risultati

*file di sistema  -> .sys -> file binario copia del file attivo resgitrato sul disco per essere usato in altre sessioni


STRUTTURA FINESTRA

Righe -> casi

Colonne -> variabili

La barra di stato (parte inferiore della finestra di SPSS, riportata nella figura successiva)
fornisce, invece, le seguenti informazioni:
1. stato di esecuzione dei comandi – per ogni procedura o comando eseguito sarà indicato il
numero di casi elaborati. Per le procedure statistiche che richiedono un'elaborazione
iterativa sarà visualizzato il numero di iterazioni

2. stato del filtro – se è stato selezionato un campione casuale o un sottoinsieme di casi per
l'analisi, il messaggio indica che è attivo un filtro e che non tutti i casi del file di dati sono
inclusi nell'analisi

3. stato della ponderazione – indica la presenza di una variabile di ponderazione usata per
pesare i casi presenti nella matrice di dati

4. stato del comando – indica che i casi della matrice dei dati sono stati divisi in gruppi distinti
secondo i valori di uno più variabili di selezione.
