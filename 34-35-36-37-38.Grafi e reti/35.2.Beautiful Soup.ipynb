BEAUTIFUL SOAP 

```python
#Beautiful Soap è una libreria per estrarre dati da file HTML e XML
    #essa controlla i tag HTML
from bs4 import BeautifulSoup

#Codice HTML di esempio
html_doc="""<html>
    <title>Mio Testo</title>
        <body>
            <h1 id='titolo'>Titolo</h1>
            <p class='pippo' id='testo1'>Contenuto del paragrafo</p>
            <a id='link1' href="https://ubuntu.com/download/desktop">LINK</a>
            <p id='testo2'>Testo</p>
            <a class='pippo' id='link2' href="https://google.it">LINK</a>


        </body>
</html>"""

#BeautifulSoup passiamo
    #html_doc -> la stringa contenente il codice HTML
    #'html.parser' -> il parser da utilizzare
soup=BeautifulSoup(html_doc,'html.parser')
#soup contiene il risultato

print("Codice HTML -> prettify()")
#prettify() -> permette di visualizzare come i tag HTML sono annidati
print(soup.prettify())
print()
print("Testo senza tag -> getText())")
print(soup.getText())

print("NAVIGARE NELLA STRUTTURA DATI")
#Per navigare nella struttura dati è possibile usare

print("->Tramite le proprietà")
    #PROPRIETA'
print("Tag title -> title")
        #title -> tag title
print(soup.title)

print("Nome del tag title -> title.name")
        #title.name -> nome del tag title
print(soup.title.name)

print("Contenuto del tag title -> title.name.string")
        #title.name.string -> il contenuto del tag title
print(soup.title.string)

print("Tag padre del tag title -> title.parent.name")
        #soup.title.parent.name -> tag padre del tag title
print(soup.title.parent.name)

print("Primo <p> che incontra -> p")
        #p -> estrae il primo <p> che incontra
print(soup.p)

print("Classe del primo <p> che incontra-> p['class']")
        #p['class'] -> estrae la classe del primo <p> che incontra
print(soup.p['class'])

print("Primo <a> che trova -> a")
        #a-> il primo <a> che trova
print(soup.a)

print("->Tramite i metodi")
    #METODI
print("Tutti i tag <a> -> findall(a)")
        #findall(t) -> estrae tutti i tag <t>
print(soup.find_all('a'))

print("Tutti i tag con id=link2 -> find(link2)")
        #find(id) -> estrae tutti i tag con id
print(soup.find(id='link2'))
```

ESTRARRE DATI DAL TAG TABLE

STAMPA NORMALE

```python
from bs4 import BeautifulSoup
import urllib.request 


URL="https://en.wikipedia.org/wiki/List_of_countries_and_territories_by_number_of_land_borders"
  
with urllib.request.urlopen(URL) as source:
  soup=BeautifulSoup(source)

#Estraggo
  # "(table).....[0]" -> la prima table che trova
    #{"class":"wikitable sortable"} -> con classe wikitable sortable
table=soup.findAll("table",{"class":"wikitable sortable"})[0]

#dizionario che conterrà uno stato e i suoi confinanti

#print(table) -> stampa tutta la <table>
#Estraggo tutte le <tr> che trova dentro la <table>
for row in table.findAll('tr'):
  #riga per riga trova tutti i <td>
  row=row.findAll('td')
  #print(row) -> stampa tutti i td trovati
  
  #se la riga ha un contenuto...
  if row:
    #estrai il primo elemento -> lo stato
    countryA=list(row[0].strings)[1] #tratta il dato come stringa
    #print(countryA) -> stampa lo stato trovato
    
    print(countryA," ---> ", end="") #usato per stampare gli stati confinati
    
    #prendi le a della sesta colonna -> stati confinanti  
    B=row[5].findAll('a') #sono tutti dei link
    
    for x in B: #scrorre gli stati confinanti trovati
      if("[" not in x.string): #cerca quelle senza il carattere "["
        print(x.string+"\t",end="") #stampa gli stati in questione
    print()
  
```

USO DI UN DIZIONARIO PREVIA LISTA CHIAVE VALORE

```python
from bs4 import BeautifulSoup
import urllib.request 

URL="https://en.wikipedia.org/wiki/List_of_countries_and_territories_by_number_of_land_borders"
  
with urllib.request.urlopen(URL) as source:
  soup=BeautifulSoup(source)

table=soup.findAll("table",{"class":"wikitable sortable"})[0]

#dizionario che conterrà uno stato e i suoi confinanti
dizionario_paesi=dict()

#Estraggo tutte le <tr> che trova dentro la <table>
  #e prendo dal l'indice 1 in poi
for row in table.findAll('tr')[1:]:
  row=row.findAll('td')
  
  if row:
    countryA=list(row[0].strings)[1]
    B=row[5].findAll('a') #sono tutti dei link
    
    # lista conterrà gli stati confinanti
    countryB_list=list() #viene ogni volta svuotata
    
    for x in B: 
      if("[" not in x.string): 
        countryB_list.append(x.string)
        
    #nella posizione dello stato attuale aggiunge via via gli stati confinanti
    dizionario_paesi[countryA]=countryB_list
    
print(dizionario_paesi) #stampa il dizionario
```

USO DI UNA SERIES A PARTIRE DA UN DIZIONARIO

```python
from bs4 import BeautifulSoup
import urllib.request 
import pandas as pd

URL="https://en.wikipedia.org/wiki/List_of_countries_and_territories_by_number_of_land_borders"
  
with urllib.request.urlopen(URL) as source:
  soup=BeautifulSoup(source)

table=soup.findAll("table",{"class":"wikitable sortable"})[0]

#dizionario che conterrà uno stato e i suoi confinanti
dizionario_paesi=dict()


for row in table.findAll('tr')[1:]:
  row=row.findAll('td')
  
  if row:
    countryA=list(row[0].strings)[1]
    B=row[5].findAll('a') #sono tutti dei link
    
    countryB_list=list()
    
    for x in B: 
      if("[" not in x.string): 
        countryB_list.append(x.string)
        
    #nella posizione dello stato attuale aggiunge via via gli stati confinanti
      #la chiave è lo stato e i suoi valori sono i confinanti
    dizionario_paesi[countryA]=countryB_list
  
  serie_paesi=pd.Series(dizionario_paesi)
  print(serie_paesi)
```

ESTRAZIONE KILOMETRI

```python
from bs4 import BeautifulSoup
import urllib.request 
import pandas as pd

URL="https://en.wikipedia.org/wiki/List_of_countries_and_territories_by_number_of_land_borders"
  
with urllib.request.urlopen(URL) as source:
  soup=BeautifulSoup(source)

table=soup.findAll("table",{"class":"wikitable sortable"})[0]

dizionario_paesi=dict()

#Contiene la lunghezza del confine
dizionario_lunghezze=dict()

for row in table.findAll('tr')[1:]:
  row=row.findAll('td')
  
  if row:
    countryA=list(row[0].strings)[1]
    
    try:
        #rimuove le vigole di separazione
        lenght=float(row[1].text.replace(",",""))
    except:
        #se non ha lunghezza mette 0
        lenght=0.0
        
    B=row[5].findAll('a') 
  
    countryB_list=list()
    
    for x in B: 
      if("[" not in x.string): 
        countryB_list.append(x.string)
    
    #la chiave è lo stato e i suoi valori sono i confinanti
    dizionario_paesi[countryA]=countryB_list
    
    #la chiave è lo stato e i suoi valori le lunghezze
    dizionario_lunghezze[countryA]=lenght
  
print(dizionario_lunghezze)
```
