GRAFI PERSONALIZZATI e OPZIONI AGGIUNTIVE

```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt

grafo=nx.Graph()

#A partire dagli archi, si creeranno anche i nodi
grafo.add_edge(1,2)
grafo.add_edge(2,3)
grafo.add_edge(3,4)
grafo.add_edge(1,4)
grafo.add_edge(1,5)

#OPZIONI AGGIUNTIVE -> Dizionario chiave/valore
    #chiave -> nome opzione
    #valore ->settaggio impostato
opzioni={
    "font_size":36,
    "node_size":3000,
    "node_color": "white",
    "edgecolors":"black",
    "linewidths":5,
    "width":5,
    }
#POSIZIONI PERSONALIZATE -> Dizionario chiave/valore
    #chiave -> nodo
    #valore -> tupla di due elementi -> coordinate xy
posizioni={1: (0,0), 2:(-1,0.3), 3:(2,0.17),4:(4,0.255),5:(5,0.03)}

#Disegna un grafo usando:
    # posizioni 
    # opzioni -> spacchetta il dizionario
print("GRAFO PERSONALIZZATO ->draw_networkx(grafo, posizioni,**option)")
nx.draw_networkx(grafo, posizioni,**option)
#MARGINI per gli assi per non cancellare parte dei nodi
ax=plt.gca() 
ax.margins(0.20)
plt.axis('off')
plt.show()

print("GRAFO PER COLONNE ->draw_networkx(grafo, posizioni,**option)")
#Grafo orientato
grafo=nx.DiGraph([(0,3),(1,3),(2,4),(3,5),(3,6),(4,6),(5,6)])
#Indichiamo quali nodi in delle LISTE dovranno stare a sinistra, centro e destra
nodi_sinistra=[0,1,2]
nodi_centrali=[3,4]
nodi_destra=[5,6]

#impostazione della coordinata x in base alla colonna
# i conterrà via via i nodi_sinistra, nodi_centrali e nodi_destra
posizioni={n:(0,i) for i,n in enumerate(nodi_sinistra)} #coordinata x di quelli a sinistra più piccola
posizioni.update({n:(1,i+0.5) for i,n in enumerate(nodi_centrali)}) #coordinata x di quelli a centra più grande
posizioni.update({n:(2,i+0.5) for i,n in enumerate(nodi_destra)}) #coordinata x di quelli a destra ancora più grande

nx.draw_networkx(grafo, posizioni,**opzioni)
#MARGINI per gli assi per non cancellare parte dei nodi
ax=plt.gca() 
ax.margins(0.20)
plt.axis('off')
plt.show()

```
