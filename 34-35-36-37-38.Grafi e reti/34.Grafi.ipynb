Una RETE  una collezione di oggetti interconnessi tra loro.

La rappresentazionione di una rete avviene sottoforma di GRAFO.

I NODI(vertici) sono collegati tra loro tramite gli ARCHI(spigoli)



V -> insieme dei nodi

E -> insieme degli archi

Un grafo  una coppia 

G=(V,E)

Due nodi u,v connessi da un arco sono l'estremo dell'arco

L'arco viene identificato dalla coppia u,v suoi estremi




Un grafo  ORIENTATO(o diretto) se almeno un arco connette il nodo A al noto B ma non il viceversa, altrimenti  un grafo NON ORIENTATO(o indiretto)

Grafo con archi paralleli, da A a B e da B ad A-> MULTIGRAFO



Un nodo con due estremi coincidenti -> CAPPIO
Un grafo senza cappi -> GRAFO SEMPLICE

Ad un grafo possono essere assegnati dei PESI -> GRAFO PESATO



Il peso  dato da un numero reale ma anche da razionali, e se i nodi non sono connessi, il loro arco peso  INFINITO
Un peso pu indicare la qualit della connessione o la distanza tra due nodi

Il GRADO DI UN NODO -> numero di archi connessi(incidenti) ad in nodo
GRADO IN INGRESSO -> Archi che entrano nel nodo
GRADO DI USCITA -> Archi che escono dal nodo



DENSIT DEL GRAFO -> Quando questo si avvicina ad un grafo completo



K indica il grado di connessione tra i nodi

GRADO MEDIO DEI NODI -> Dimensione Grafo(numero di archi)/Ordine del grafo(numero di nodi)

GRAFO COMPLETO -> un grafo che contiene tutti gli archi possibili, ovvero quando tutti i nodi sono connessi tra loro
Densit per un grafo orientato



dove 
e-> numero di archi
n-> numero di nodi
Densit per un grafo non orientato



GRADO MASSIMO -> Grado del vertice con il maggior numero di archi incidenti

GRADO MINIMO -> Grado del vertice con il minor numero di archi

Se grado massimo e minimo coincidono -> GRAFO REGOLARE (o K-REGOLARE)

ORDINE di un grafo -> numero di vertici

DIMENSIONE di un grafo -> Numero di archi

GRAFO NULLO -> Grafo primo di archi




STRUTTURA DI UN GRAFO
La connettivit dei nodi un grafo e dati dai cammini
CAMMINO -> Qualsiasi sequenza di archi tale per cui la fine di un arco  l'inizio di un altro arco



PERCORSO (o path) ->  un cammino che non si interseca mai -> non incontra pi volte lo stesso nodo

Un percorso DI LUNGHEZZA n del grafo g  dato da:

*SEQUENZA DI VERTICI V0,V1,...,Vn

*SEQUENZA DI ARCHI che li collegano (V0,V1), (V1,V2),...,(Vn-1,Vn)

PERCORSO CHIUSO -> Nodo iniziale e finale uguali -> CICLO



Percorso H-A-B in verde

Ciclo senza vertici o archi ripetuti H-D-G-H in rosso

Un cammino con vertici ripetuti B-D-E-F-D-C-B in blu

DISTANZA FRA DUE NODI -> Percorso pi piccolo tra due archi di ogni percorso. Questo concetto non esiste negli archi pesati

In un grafo orientato la DISTANZA TRA A e B non  sempre uguale alla DISTANZA TRA B e A

DIAMETRO D del grafo -> massima distanza fra due nodi qualsiasi


COMPONENTE CONNESSO -> insieme dei nodi di un grafo per cui esiste un percorso da ogni nodo dell'insieme a ogni altro nodo dell'insieme stesso




In un grafico orientato avremo

*COMPONENTI FORTEMENTE CONNESSI -> connessi da efettivi percorsi
 
*COMPONENTI DEBOLMENTE CONNESSI -> connessi da percorsi i cui archi sono stati converiti in archi non orientati

CENTRALIT  AD UN AUTOVETTORE A  definita ricorsivamente come la somma scalare delle centralit ad autovettore di tutti i nodi adiacenti di A. Essa prende in considerazione la connettivit con i vicini a livello globale. I punteggi relativi vengono assegnati a tutti i nodi della rete in base al concetto che le CONNESSIONI AI NODI CON IL PUNTEGGIO PIU' ELEVATO contribuiscono maggiormente al punteggio del nodo in questione rispetto alle stesse connessioni, ma a nodi con punteggio basso. La centralit ad autovettore , quindi, una misura dell'INFLUENZA DI UN NODO nella rete.


Centralit, centralit globale dei flussi e centralit ad autovettore, sono costose dal punto di vista computazionale e pu essere poco pratico calcolarle per reti di grandi dimensioni.
