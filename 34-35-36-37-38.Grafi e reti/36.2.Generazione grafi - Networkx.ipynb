```python

import networkx as nx
import math

#Grafo vuoto
grafo=nx.Graph()

print("NODI")

print("Aggiunta nodo -> add.node(1)")
#aggiunta di un nodo 1
grafo.add_node(1)

print("Aggiunta nodi -> add_nodes_from([2,3])")
#aggiungo i nodi 2 e 3
grafo.add_nodes_from([2,3])


print("Aggiunta nodi da container ->path_graph(5) + add_nodes_from(container_nodi)")
#container di 5 nodi -> da 0 a 4
container_nodi=nx.path_graph(5)
#aggiungo da container
grafo.add_nodes_from(container_nodi)

print("Rimozione nodo -> remove_node(2)")
#rimuovo il nodo 2
grafo.remove_node(2)

print("Aggiunta di nodi di altri tipi -> add.node('stringa' O Funzione O Oggetto...)")
#ai nodi posso aggiungere oggetti, stringhe, funzioni, filtri..
#aggiungo un nodo stringa
grafo.add_node('stringa')

#aggiungo un nodo funzione
grafo.add_node(math.cos)

file=open('temp.txt','w')
#aggiungo un nodo file
grafo.add_node(file)

print("Mostrare nodi -> nodes()")
#mostra tutti i nodi
print(grafo.nodes())

print("ARCHI")

print("Aggiungere arco -> add_edge(1,2)")
#aggiungere un arco tra il nodo 1 e 2   
    #Se manca il nodo, questo verrà aggiunto
grafo.add_edge(1,2)

print("Aggiungere archi da tupla-> arco=(2,3) e add_edge(*arco)")
#arco come tupla
arco=(2,3)
#aggingo un arco come tupla
    # * -> spacchetta la tupla
grafo.add_edge(*arco)

print("Aggiungere archi da lista di tuple -> add_edges_from([(1,2),(1,3)]) e")
#lista di tuple di archi aggiunti al grafo
grafo.add_edges_from([(1,2),(1,3)])
#oppure
print("oppure -> container_archi=[(1,2),(1,3)]) e grafo.add_edges_from(container_archi)")
container_archi=[(1,2),(1,3)]
grafo.add_edges_from(container_archi)

print("Rimuovere arco -> remove_edge(1,2)")
#rimuovere arco
    #arco tra 1 e 2
grafo.remove_edge(1,2)

#Mostra gli archi
print("Mostrare archi -> edges()")
print(grafo.edges())

print("DATI SUL GRAFO")
print("Ordine grafo (numero di nodi) -> number_of_nodes()) o order() o len(grafo)")
#Ordine del grafo(numero di nodi) -> equivalenti
    #metodi del grafo
print(grafo.number_of_nodes())
print(grafo.order())
    #metodo a cui passiamo il grafo
print(len(grafo))

print("Dimensione grafo (numero di archi) -> grafo.number_of_edges() o size()")
#Dimensione del grafo (numero di archi) -> equivalenti
print(grafo.number_of_edges())
print(grafo.size())

print("Vicini di un nodo (nodi adiacenti o legati al nodo)-> neighbors(1)")
#vicini del nodo 1
vicini=grafo.neighbors(1)
for i in vicini:
    print(i)

print("Grado di un nodo (quanti vicini ha un nodo) -> degree(1)")
#grado di un nodo
print(grafo.degree(1))
```

AGGIUNTA NODI CHIAVE VALORE -> DIZIONARIO e PESO

```python
import networkx as nx
import math

#Grafo vuoto
grafo=nx.Graph()
print("Ripasso dizionario")
#dizionario non omogeneo
dizionario={'apple':1, 'orange': [0.12,0.02], 42:21}
    #stringa : numero
    #stringa : lista
    #numero : booleano
print("Chiavi e valori")
print(dizionario.keys()) #chiavi
print(dizionario.items()) #valori -> tuple

print('NODO COME DIZIONARIO')
#un nodo si comporta come un dizionario Python
grafo.add_node(1,time='10am')
    #Aggiungo un nodo 1 con attributo time='10am'
    
print('Accedo all\'attributo del nodo 1 con chiave \'time\' -> nodes[1][\'time\']')
#prendo il nodo [1] e il valore di ['time']
print(grafo.nodes[1]['time']) # -> 10am

print('Tipo nodo')
d=grafo.nodes[1] #tutto il nodo 1
#ogni singolo nodo è un DIZIONARIO
print(type(d)) #dict_keys
print('Contenuto nodo')
print(d) #{'time': '10am'}

print('PESO ARCO')
#Peso per i grafi pesati tramite weight

print('Peso arco -> add_edge(1,2, weight=4.0)')
    #aggiungo un arco tra 1 e 2 con peso a 4
grafo.add_edge(1,2, weight=4.0)
print(grafo[1][2])
print('Modifica peso arco -> grafo[1][2][\'weight\']=5.0)')

#modifico il peso a 5
grafo[1][2]['weight']=5.0
print(grafo[1][2])


print("ITERAZIONE GRAFO")

print("Scorro i nodi -> for nodo in grafo.nodes()")
for nodo in grafo.nodes():
    print("Nodo : ",nodo, " Grado : ",grafo.degree(nodo))


print("Scorro gli archi -> for arco in grafo.edges()")
for arco in grafo.edges():
    print("Arco ",arco)
    
print("Scorro gli archi per nodi e peso -> for nodo1,nodo2,peso in grafo.edges(data=True):")

#per ongi arco accedo ai dati di ciascuno -> data=True
    #in ordine accedo al nodo1, nodo2, peso 
        #grafo.add_edge(1,2, weight=4.0) -> compaiono in qusto ordine
for nodo1,nodo2,peso in grafo.edges(data=True):
    print("Nodo 1 : ",nodo1, "Nodo 2 : ",nodo2," Peso : ",peso)
```

GRAFO ORIENTATO -> DiGraph

```python
import networkx as nx
import math

#Grafo ORIENTATO vuoto
grafo_orientato=nx.DiGraph() #DirectiveGraph

#Aggiungo al grafo 2 nodi:
    #da 1 a 4 con peso 0.5
    #da 3 a 1 con peso 0.75
grafo_orientato.add_weighted_edges_from([(1,4,0.5),(3,1,0.75)])
#(3)--0.5-->(1)--0.75-->(4)


print("Grado in uscita dal nodo 1 : ",grafo_orientato.out_degree(1))
print("Grado del nodo 1: ",grafo_orientato.degree(1))

print("Peso in uscita dal nodo 1 : ",grafo_orientato.out_degree(1,weight='weight'))
print("Peso del nodo 1: ",grafo_orientato.degree(1,weight='weight'))

#I successori sono i nodi vicini dal nodo da un arco uscente
successori=grafo_orientato.successors(1)
print("Successore/i del nodo 1: ")
for n in successori:
    print(n)
    
#I successori sono i nodi vicini dal nodo da un arco entrante
predecessori=grafo_orientato.predecessors(1)
print("Predecessore/i del nodo 1: ")
for n in predecessori:
    print(n)
```
