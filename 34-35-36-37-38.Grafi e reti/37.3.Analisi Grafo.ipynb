GRAFI PERSONALIZZATI e OPZIONI AGGIUNTIVE

```python
%matplotlib inline
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

grafo=nx.Graph()

#A partire dagli archi, si creeranno anche i nodi
grafo.add_edge(1,2)
grafo.add_edge(2,3)
grafo.add_edge(3,4)
grafo.add_edge(1,4)
grafo.add_edge(1,5)

#OPZIONI AGGIUNTIVE -> Dizionario chiave/valore
    #chiave -> nome opzione
    #valore ->settaggio impostato
opzioni={
    "font_size":36,
    "node_size":3000,
    "node_color": "white",
    "edgecolors":"black",
    "linewidths":5,
    "width":5,
    }
#POSIZIONI PERSONALIZATE -> Dizionario chiave/valore
    #chiave -> nodo
    #valore -> tupla di due elementi -> coordinate xy
posizioni={1: (0,0), 2:(-1,0.3), 3:(2,0.17),4:(4,0.255),5:(5,0.03)}

print("GRAFO PER COLONNE ->draw_networkx(grafo, posizioni,**option)")
#Grafo orientato
grafo=nx.DiGraph([(0,3),(1,3),(2,4),(3,5),(3,6),(4,6),(5,6)])
#Indichiamo quali nodi in delle LISTE dovranno stare a sinistra, centro e destra
nodi_sinistra=[0,1,2]
nodi_centrali=[3,4]
nodi_destra=[5,6]

#impostazione della coordinata x in base alla colonna
# i conterr via via i nodi_sinistra, nodi_centrali e nodi_destra
posizioni={n:(0,i) for i,n in enumerate(nodi_sinistra)} #coordinata x di quelli a sinistra pi piccola
posizioni.update({n:(1,i+0.5) for i,n in enumerate(nodi_centrali)}) #coordinata x di quelli a centra pi grande
posizioni.update({n:(2,i+0.5) for i,n in enumerate(nodi_destra)}) #coordinata x di quelli a destra ancora pi grande

nx.draw_networkx(grafo, posizioni,**opzioni)
#MARGINI per gli assi per non cancellare parte dei nodi
ax=plt.gca() 
ax.margins(0.20)
plt.axis('off')
plt.show()


print("ANALISI DEI GRAFI")

#Ordine grafo
print("Numero dei nodi : ",len(grafo)) # oppure print(grafo.number_of_nodes()) o print(grafo.order())

#Dimensione grafo
print("Numero degli archi : ",grafo.size()) # oppure print(grafo.number_of_edges())

print("Grado medio dei nodi : ",float(grafo.size()/len(grafo)))

# connessi da effettivi percorsi
print("Numero di componenti fortemente connessi : ",nx.number_strongly_connected_components(grafo))

# connessi da percorsi i cui archi sono stati converiti in archi non orientati
print("Numero di componenti debolmente connessi : ",nx.number_weakly_connected_components(grafo))

print("Componenti connesse -> SOLO SU GRAFI NON ORIENTATI -> nx.connected_components(nx.to_undirected(grafo)") 
#Per il calcolo delle componenti connesse 
# SOLO SU GRAFI NON ORIENTATI
#conto le componenti sul grafo non orientato
componenti=nx.connected_components(nx.to_undirected(grafo))

c=0
for i in componenti:
    print(i)
    c+=1
print(c," componenti connesse")

print("GRADO USCENTE ED ENTRANTE -> SOLO GRAFI ORIENTATI")
#SERVE UN GRAFO ORIENTATO!
#Nei grafi orientati  possibile calcolare il grado entrante o uscente
print("Grado entrante uscente -> in_degree()")
lista_in=list()
gradi_entrante=grafo.in_degree() #ritorna tuple -> Tupla(nodo, grado entrante)
for n in gradi_entrante:
    lista_in.append(n[1]) #il secondo valore  il grado entrante
print(lista_in)
#istogramma
plt.hist(lista_in,bins=np.arange(-2,5)+0.5,ec='k')
plt.show()

print("Grado uscente -> out_degree()")
lista_out=list()
gradi_uscente=grafo.out_degree() #ritorna tuple -> Tupla(nodo, grado uscente)
for n in gradi_uscente:
    lista_out.append(n[1]) #il secondo valore  il grado uscente
print(lista_out)
#istogramma
plt.hist(lista_out,bins=np.arange(-2,5)+0.5,ec='k')
plt.show()
```

```python
%matplotlib inline
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

grafo=nx.Graph()

#A partire dagli archi, si creeranno anche i nodi
grafo.add_edge(1,2)
grafo.add_edge(2,3)
grafo.add_edge(3,4)
grafo.add_edge(1,4)
grafo.add_edge(1,5)

#OPZIONI AGGIUNTIVE -> Dizionario chiave/valore
    #chiave -> nome opzione
    #valore ->settaggio impostato
opzioni={
    "font_size":36,
    "node_size":3000,
    "node_color": "white",
    "edgecolors":"black",
    "linewidths":5,
    "width":5,
    }
#POSIZIONI PERSONALIZATE -> Dizionario chiave/valore
    #chiave -> nodo
    #valore -> tupla di due elementi -> coordinate xy
posizioni={1: (0,0), 2:(-1,0.3), 3:(2,0.17),4:(4,0.255),5:(5,0.03)}

print("GRAFO PER COLONNE ->draw_networkx(grafo, posizioni,**option)")
#Grafo orientato
grafo=nx.DiGraph([(0,3),(1,3),(2,4),(3,5),(3,6),(4,6),(5,6)])
#Indichiamo quali nodi in delle LISTE dovranno stare a sinistra, centro e destra
nodi_sinistra=[0,1,2]
nodi_centrali=[3,4]
nodi_destra=[5,6]

#impostazione della coordinata x in base alla colonna
# i conterr via via i nodi_sinistra, nodi_centrali e nodi_destra
posizioni={n:(0,i) for i,n in enumerate(nodi_sinistra)} #coordinata x di quelli a sinistra pi piccola
posizioni.update({n:(1,i+0.5) for i,n in enumerate(nodi_centrali)}) #coordinata x di quelli a centra pi grande
posizioni.update({n:(2,i+0.5) for i,n in enumerate(nodi_destra)}) #coordinata x di quelli a destra ancora pi grande

nx.draw_networkx(grafo, posizioni,**opzioni)
#MARGINI per gli assi per non cancellare parte dei nodi
ax=plt.gca() 
ax.margins(0.20)
plt.axis('off')
plt.show()


print("ANALISI DEI GRAFI")

print("COEFFICENTE DI CLUSTERING LOCALE -> SOLO GRAFI NON ORIENTATI")
#SERVE UN GRAFO NON ORIENTATO!
grafo_non_orientato=grafo.to_undirected()
#Numero di archi nel vicinato di A/massimo numero possibile di archi
    #NetworkX calcola il coefficente di clustering:
        #riferito ad un nodo
        #riferito a tutti i nodi
        #medio riferito al grafo
        
print("Coefficente di clustering di un nodo ->clustering(grafo_non_orientato,nodo)")
print("nodo 0 -> ",nx.clustering(grafo_non_orientato,0))

print("Coefficente di clustering di tutti i nodi ->clustering(grafo_non_orientato)")
#come quello di prima ma passo un solo parametro
clust_coefficenti=nx.clustering(grafo_non_orientato) # un dizionario

print("Coefficenti di clustering dei nodi:")
for i in clust_coefficenti:
    print("Nodo ",i," con coefficente di clustering ","{:3.4f}".format(clust_coefficenti[i]))

print("Media coefficente di clustering: -> nx.average_clustering(grafo_non_orientato)")
print("{:3.4f}".format(nx.average_clustering(grafo_non_orientato)))
```
