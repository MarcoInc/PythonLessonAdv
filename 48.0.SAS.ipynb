```python
SAS -> Statistical Analysis System

Complesso di software che permette di:

    *Inserimento,ricerca e gestione di dati

    *generazione di report e grafici

    *analisi statistica e matematica

    *pianificazione, previsione e supporto alle decisioni

    *ricerca operativa e project management

    *gestione di qualità

    *sviluppo di applicazioni

In nucleo di SAS è basato sulle applicazioni:

*Base SAS software -> utilizzato per la gestione dei dati

*SAS procedure software -> analisi e reporting

*Macro facility -> personalizzare le applicazioni

*DATA step debugger -> tool per individuare gli eventuali problemi nelle applicazioni sviluppate

*Output Delivery System (ODS) -> modulo che tratta risultato per restituirlo
in formati standard e facilmente gestibili -> SAS DATASET, listening e HTML

*SAS windowing enviroment -> interfaccia grafica e interattiva pereseguire e testare le 
applicazioni sviluppate nell'ambiente SAS


AREE FINESTRA DI SAS

    *EDITOR -> scrittura codice

    *LOG -> Vista messaggi

    *RISULTATI -> risulati

TABELLA CLASS -> Tabella con tutte le osservazioni in riga e variabili in colonna

COMANDI SAS

    *KEYWOD SAS -> PROC, PUT, DATA, FORMAT

    *NOMI SAS -> variabili, archivi, procedure e zioni

    *CARATTERI SPECIALI ed OPERATORI -> ; = $ > + / > * : ^ ()



Un programma SAS si articola in uno più step

    *DATA STEP -> crea e gestisce i DATASET -> archivi organizzati
        
        Inizia sempre con l'istruzione DATA

            DATA x; -> x è il nome dell'archivio

        Tuttavia non sempre crea DATASET


    *PROC STEP -> serve a richiamare una procedura SAS da applicare ai 

SAS DATASET ->-DATASET AUTODESCRITTIVO -> contiene dati e metadati(info sui dati)
     
     non è un file di testo AUTODESCRITTIVI


NOME VARIABILI

    Non c'è distinzione tra maiuscole e minuscole     

    Max 32 caratteri 

    Iniziano solamente con lettera o _ (uderscore)

    Fine -> nessun carattere numerico

TIPO VARIABILE -> possono avere un'etichetta da stringa

    *Numeriche -> da 2 ad 8 byte
        
        solo da numeri, o decimali

    *Carattere -> 1 a 32767 byte

    
    *Altritipi -> derivati dai precedenti (booleani)

Tutti gli step fanno uso di ISTRUZIONI:

    data statement

    set statement

    assigment statement

    run statement

Il primo passo per scrivere codice SAS è creare un archivio SAS

Per operare su un archivio si usa DATA

    DATA x; -> x è il nome dell'archivio
```
