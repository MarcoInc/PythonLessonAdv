```python
import findspark 
findspark.init()

import pyspark
#Per usare i dataframe si usa SparkSession contenuto in pyspark.sql
from pyspark.sql import SparkSession #SparkContext è contentuto in SparkSession
from pyspark import SparkContext #usato successivamente per creare il RDD

from pyspark.sql import functions as F 

#TRA RDD E DataFrame ci sono due generazioni diverse
spark=SparkSession.builder.getOrCreate()
dati=[('Mario','Rossi',1.7,43,'1999-12-25','M'),
('Carla','Verdi',1.5,54,'1990-08-09','F'),
('Marta','Bianchi',1.75,82,'2000-11-28','F'),
('Pippo','Neri',1.65,77,'2001-10-07','M')]

#creo uno sparkContext
sc=spark.sparkContext

rdd=sc.parallelize(dati)
#Creare un DataFrame da un RDD specificando:
    #RDD di partenza
    #Nome colonne
df=spark.createDataFrame(rdd,['nome','cognome',
                                    'altezza','peso','nascita','sesso'])
df.show()

#I DATI VANNO PRESENTATI NEL MODO MIGLIORE
    #orderBy -> ritorna un altro Dataframe ordinato in base ad un campo specifico
    #limit() -> ritorna un altro DataFrame
        #  specifica il numero di dati che devono essere espressi
        #diverso da first, take o head -> ritornano ROW
    #select() -> definisce le proiezioni dei dati ovve
        #il modo con cui i dati devono essere restituiti
print("Ordine in base all'altezza -> orderBy(..)")
df.orderBy('altezza').show()
print("Ordine in base all'altezza DECRESCENTE")
df.orderBy('altezza', ascending=False).show()


print("Limitare la vista -> limit() + orderBy(..)")
# limit(1) -> mostra un solo me
df.orderBy('altezza').limit(1).show()

print("Selezionare i dati da mostrare -> select()")
#Raggrupa per sesso
#Calcola il massimo e minimo
#When controla il sesso
    #se è M mette MASCHIO
    #otherwise-> altrimenti FEMMINA
#alias -> rinomina le colonne rispettive
df.groupBy('sesso')\
.agg(F.min('peso').alias('peso minimo'),F.max('peso').alias('peso massimo'))\
.select(F.when(df.sesso=='M','MASCHIO').otherwise('FEMMINA')\
        .alias('sesso'),'peso minimo','peso massimo').show()
```
