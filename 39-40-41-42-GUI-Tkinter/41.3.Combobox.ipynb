ttk.Combobox

widget ttk.Combobox è un widget Entry che aggiunge un menu di selezione drop-down.
Per popolare il menu drop-down è necessario passare all'argomento values una lista di
stringhe, tra cui l'utente può scegliere.





• parent è l'argomento che imposta il widget parent (gerarchicamente superiore) che
conterrà il widget Combobox;

• textvariable è un oggetto di tipo StringVar di Tkinter il cui valore sarà collegato a questo
widget di input;

• values, come anticipato, è una lista di stringhe che rappresentano le opzioni, tra cui
l’utente può scegliere.
