ttk.Spinbox

Il widget ttk.Spinbox aggiunge due pulsanti (uno per l'incremento e uno per il decremento)
ad un normale widget Entry rendendolo adatto a dati numerici e anche lista di stringhe.





• parent è l'argomento che imposta il widget parent (gerarchicamente superiore) che
conterrà il widget Spinbox;

• from_ (obbligatorio se to è presente) è l'argomento che determina il minimo valore esprimibile dal widget. Il trattino basso
è necessario perché from è una keyword Python;

• to (ibbligatorio se from_ è usato)  è l'argomento che determina il massimo valore esprimibile dal widget;

• increment è l'argomento che rappresenta la quantità di incremento/decremento;

• textvariable è un oggetto StringVar di Tkinter il cui valore sarà collegato a questo widget di
input.
