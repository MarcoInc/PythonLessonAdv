ttk.Spinbox

Il widget ttk.Spinbox aggiunge due pulsanti (uno per l'incremento e uno per il decremento)
ad un normale widget Entry rendendolo adatto a dati numerici e anche lista di stringhe.





 parent  l'argomento che imposta il widget parent (gerarchicamente superiore) che
conterr il widget Spinbox;

 from_ (obbligatorio se to  presente)  l'argomento che determina il minimo valore esprimibile dal widget. Il trattino basso
 necessario perch from  una keyword Python;

 to (ibbligatorio se from_  usato)   l'argomento che determina il massimo valore esprimibile dal widget;

 increment  l'argomento che rappresenta la quantit di incremento/decremento;

 textvariable  un oggetto StringVar di Tkinter il cui valore sar collegato a questo widget di
input.
