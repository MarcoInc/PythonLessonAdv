ttk.Button

Il widget ttk.Button è un pulsante cliccabile con il mouse o lo spazio della tastiera





Come il widget Checkbutton, questo widget usa le opzioni di configurazioni text e
textvariable per controllare l'etichetta sul pulsante. L'argomento command, invece, specifica una
funzione Python (funzione di callback) da invocare quando il pulsante viene cliccato.
