delete(), altrimenti si richiamano i metodi delete() e insert() dei widget di input che sono invocabili
con le classi Entry, Spinbox e Combobox.

quello del widget di tipo tk.Text perché l’indicizzazione funziona in modo differente.

Ovviamente la classe LabelInput per come è stata scritta non gestisce tutti i possibili widget
di input, ma copre quelli che verranno utilizzati per il form.

La costruzione della classe LabelInput ha richiesto molto lavoro, ma grazie ad esso,
l'implementazione del form sarà molto più semplice.
