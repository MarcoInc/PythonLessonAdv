Il costruttore della nuova classe riceve i seguenti argomenti:



• parent è il riferimento al widget parent che conterrà il widget LabelInput;

• label è il testo delle etichette del nuovo widget;

• input_class è la classe del widget che si vuole creare. Deve essere un oggetto input, non
una stringa. Se lasciato vuoto, il default del tipo di input è impostato come widget ttk.Entry;

• input_var è una variabile Tkinter a cui assegnare l'input. È opzionale in quanto alcuni widget
di input non usano le variabili;

• input_args è un dizionario opzionale per argomenti aggiuntivi per il costruttore del widget di
input;

• label_args è un dizionario opzionale per argomenti aggiuntivi per il costruttore del widget
Label;

• **kwargs serve per raccogliere argomenti keyword aggiuntivi da passare al costruttore del
Frame.

La prima istruzione all'interno del costruttore sarà la chiamata al costruttore della
superclasse frame super().__init__() passando l'argomento parent e gli argomenti keyword
aggiuntivi **kwargs.
