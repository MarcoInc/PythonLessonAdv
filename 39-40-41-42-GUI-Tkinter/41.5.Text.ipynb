tk.Text


Il widget tk.Text è un widget che è molto di più di un widget Entry multilinea. Esso fornisce,
infatti, un sistema di tag molto potente che permette di implementare testo multicolore, link
cliccabili e molto altro. A differenza degli altri widget non può essere collegato ad una StringVar
Tkinte. 

Ne segue che l’impostazione il recupero del testo contenuto dovrà essere eseguito
attraverso i metodi get(), insert() e delete().

Quando si legge o si modifica il testo con i metodi get(), insert() e delete() è necessario
passare anche i valori index per selezionare il carattere (un valore di index) o un intervallo di
caratteri (due valori di index) su cui operare. I valori di index vanno passati come stringhe secondo
uno dei seguenti formati:

1. il numero della linea e il numero del carattere separati da un punto. Le linee sono numerate
da 1, mentre i caratteri da 0. Il primo carattere della prima linea è, dunque, 1.0;

2. la stringa end o la costante END di Tkinter indica la fine del campo del widget Text;

3. un indice numerico più una delle seguenti parole linestart, lineend, wordstart, wordend
indica l'inizio o la fine della linea o della parola relativa all'indice numerico. Ad esempio, 6.2
wordstart corrisponde all'inizio della parola contenente il terzo carattere della linea 6,
mentre 2.0 lineend corrisponde alla fine della linea due.

Qualsiasi dei formati precedenti insieme a un operatore di somma/differenza e un numero
di caratteri o linee. Ad esempio, 2.5 wordend – 1 chars corrisponde al carattere prima della fine
della parola contenente il sesto carattere della linea 2.
