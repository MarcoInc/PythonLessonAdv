ttk.Checkbutton


Il widget ttk.Checkbutton  un checkbox etichettato per l'inserimento di dati booleani. A
differenza dello Spinbox e del Combobox non deriva dal widget Entry e i suoi argomenti sono
diversi





 text  l'argomento che imposta l'etichetta del widget;

 variable  l'argomento di tipo BooleanVar a cui lo stato booleano  collegato;

 textvariable a differenza del widget Entry questo argomento pu essere usato per collegare
una variabile al testo dell'etichetta del widget. Con questo widget, textvariable non viene
utilizzato molto di frequente anche se  utile sapere della sua esistenza.
