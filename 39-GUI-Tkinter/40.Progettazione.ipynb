L'implementazione di un'interfaccia grafica, necessitadi una PROGETTAZIONE

Partire dall'INPUT DEI DATI -> data entry

FORM -> scheda dell'inserimento dati

FORM permette di:    
    *inserimento dati  
    
    *integrit dati
    
    *limitare le scelte alle sole opzioni valide
    
    *autocompilazione
    
    *verificare una data sia in un intervallo atteso
        
        es-> data di nascita, prenotazione

    *verificare che tutti i dati siano inseriti

I controlli limitano gli errori

CONTROLLI:

    *valori accettabili?
    
    *campi limitati?
    
    *unita di misura?
    
    *i campi sono numeri o stringhe?
    
    *intervallo accettabile
    
    *tempo necessario per riempire il form?
    
    *formato di inserimento?
    
    *ci sono dei campi che si riempiono automaticamente?

PROGETTARE FORM:

    *determinare widget pi appropriati
    
    *raggruppare insieme di widget
    
    *organizzre layout






Per raggruppare i widget precedenti in Tkinter si pu procedere in due modi:

1. inserire semplicemente delle etichette tra ciascun insieme di campi;

2. utilizzare uno dei seguenti widget (contenitore) per raggruppare insieme al loro interno gli
altri widget:

 ttk.LabelFrame  un frame con unetichetta e un bordo opzionale;

 ttk.NoteBook  widget a schede che permettono la visualizzazione di diverse pagine;

 Tkinter.PanedWindow  permette la visualizzazione di diversi frame (ridimensionabili) disposti
in orizzontale o in verticale.

Spesso  utile realizzare anche un modello (mockup) del form per avere unanteprima
grafica dellinterfaccia (figura successiva).
