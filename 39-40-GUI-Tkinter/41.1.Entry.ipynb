ttk.Entry

Il widget ttk.Entry  un normale rettangolo di inserimento di testo su una sola linea come
mostrato nella figura seguente.





parent  l'argomento che imposta il widget parent (gerarchicamente superiore) che
conterr il widget Entry;

 textvariable  un oggetto di tipo StringVar di Tkinter il cui valore sar collegato a questo
widget di input;

textvariable  opzionale. Senza taleargomento, infatti,  possibile estrarre il valore all'interno del widget entry utilizzando il metodo get().

 show (opzionale)  l'argomento che determina quale carattere sar visualizzato quando si
inserisce il testo nel widget Entry. Di default  il carattere inserito, ma in casi speciali come per le password potrebbe essere sostituito dallasterisco come riportato nella figura
seguente.
