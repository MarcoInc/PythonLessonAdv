ttk.Entry

Il widget ttk.Entry è un normale rettangolo di inserimento di testo su una sola linea come
mostrato nella figura seguente.





•parent è l'argomento che imposta il widget parent (gerarchicamente superiore) che
conterrà il widget Entry;

• textvariable è un oggetto di tipo StringVar di Tkinter il cui valore sarà collegato a questo
widget di input;

textvariable è opzionale. Senza taleargomento, infatti, è possibile estrarre il valore all'interno del widget entry utilizzando il metodo get().

• show (opzionale) è l'argomento che determina quale carattere sarà visualizzato quando si
inserisce il testo nel widget Entry. Di default è il carattere inserito, ma in casi speciali come per le password potrebbe essere sostituito dall’asterisco come riportato nella figura
seguente.
