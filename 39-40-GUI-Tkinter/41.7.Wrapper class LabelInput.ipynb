Il costruttore della nuova classe riceve i seguenti argomenti:



 parent  il riferimento al widget parent che conterr il widget LabelInput;

 label  il testo delle etichette del nuovo widget;

 input_class  la classe del widget che si vuole creare. Deve essere un oggetto input, non
una stringa. Se lasciato vuoto, il default del tipo di input  impostato come widget ttk.Entry;

 input_var  una variabile Tkinter a cui assegnare l'input.  opzionale in quanto alcuni widget
di input non usano le variabili;

 input_args  un dizionario opzionale per argomenti aggiuntivi per il costruttore del widget di
input;

 label_args  un dizionario opzionale per argomenti aggiuntivi per il costruttore del widget
Label;

 **kwargs serve per raccogliere argomenti keyword aggiuntivi da passare al costruttore del
Frame.
