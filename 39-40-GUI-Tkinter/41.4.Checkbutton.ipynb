ttk.Checkbutton


Il widget ttk.Checkbutton è un checkbox etichettato per l'inserimento di dati booleani. A
differenza dello Spinbox e del Combobox non deriva dal widget Entry e i suoi argomenti sono
diversi





• text è l'argomento che imposta l'etichetta del widget;

• variable è l'argomento di tipo BooleanVar a cui lo stato booleano è collegato;

• textvariable a differenza del widget Entry questo argomento può essere usato per collegare
una variabile al testo dell'etichetta del widget. Con questo widget, textvariable non viene
utilizzato molto di frequente anche se è utile sapere della sua esistenza.
