ttk.Combobox

widget ttk.Combobox  un widget Entry che aggiunge un menu di selezione drop-down.
Per popolare il menu drop-down  necessario passare all'argomento values una lista di
stringhe, tra cui l'utente pu scegliere.





 parent  l'argomento che imposta il widget parent (gerarchicamente superiore) che
conterr il widget Combobox;

 textvariable  un oggetto di tipo StringVar di Tkinter il cui valore sar collegato a questo
widget di input;

 values, come anticipato,  una lista di stringhe che rappresentano le opzioni, tra cui
lutente pu scegliere.
