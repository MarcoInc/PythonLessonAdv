I dati delle osservazioni contenute in un DataSet che non hanno valori assegnati sono VALORI MANCANTI dovuti ad:

    *Mancata risposta del sistema

    *Dati danneggiati

    *Cancellazione parziale

Alcuni dati possono mancare più spesso di altri

I DATI MANCANTI possono essere un problema enorme sia nella PREPARAZIONE DEI DATI che nell'ANALISI

PRIMA di inziare l'analisi fare una percentuale dei dati mancanti

In caso di valori mancanti si procede:

    *Rimozione di interi record che mostrano valori mancanti

    *Inserimento di un valore costante al posto del valore mancante

FUNZIONI UTILI PER ANALISI DATI MANCANTI

>from pyspark.sql.functions import isnan,when,count,col

Contare gli elementi null di una colonna specifica

>dataframe_missing.filter(col('spellength').isNull()).count() 

oppure è possibile contare quanti valori mancano per ciascuna colonna

>dataframe_missing.select([count(when(isbab(c) | col(c).sNull(), c)).alias(c) for c in dataframe_missing.columns]).show

Contare i valori non mancanti in ogni colonna

>dataframe_missing.describe().show(1)

Recuperare record relativi ai dati mancanti ad una relativa colonna

>dataframe_missing.where(col('spellength').isNull()).show() 

Eliminare e contare i record con valori mancanti in una data colonna

>dataframe_missing.select('spellength').dropna().count()

Inserimento di un valore costante al posto del valore mancate di una determinata colonna -> in questo caso con 1

>dataframe_missing.select('spellength').fillna(1)

in questo caso non saranno piùù presenti valori mancanti

Inserimento di un valore costante in tutte le colonne

>dataframe_missing.s.fillna(1)
