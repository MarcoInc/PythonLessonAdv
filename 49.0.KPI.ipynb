Per analizzare un problema e definrie il suo ambito è possibile usare le METRICHE esistenti.

KEY PERFORM INDICATOR (KPI) -> Metriche che misurano il problema e illustrano i soggetti d'interesse del problema stesso

    *ogni problema specifico ne ha

    *Dopo averli generati si analizzano le tendenze e quantificare il problema

Per risolvere un problema in Python:

    1. Comprensione e definizione del problema

    2. Identificazione dei parametri chiave ad esso collegati

    3. Uso di librerie Python per il calcolo dei KPI

La risoluzione di problemi di scienza dei dati inizia definendo quantitativamente il problema e realizzando poi ANALISI e SOLUZIONI in linea con i requisiti.
Per riconoscere e definire un problema:

1. Identificazione del problema -> raccolta degli obiettivi

2. Raccolta dei requisiti -> indagare sul problema da più  punti di vista

3. Elaboranzione dei mdati e flusso di lavoro -> individuare e concordare delle metriche quantificabili. 

    *Generare quindi le KPI
    
    *Usare metriche esistente o valutare dati storici

4. Identificazioni di metriche misurabili -> sviluppare modelli descrittivi per analizzare i dati storici -> studiare il passato 

5. Documentazione e presentazione 


Per ESPLORARE I KPI:

    1. Raccolta dei dati necessari <- fanno parte del problema stesso

    2. Analisi della produzione di dati -> identificare le metriche

        -> Le metriche possono essere ottenute:

            *Manipolando i dati i dati grezzi -> estraendo quelli che ci servono

            *Disponibili nei dati forniti

        DATI NON STRUTTURATI -> Sono senza righe e colonne <- file .xml

            Vanno convertiti

        VARIABILE TARGET -> Variabile(DIPEDENTE) identificata che definisce il problema

    3. Visualizzazione dei KPI -> Rappresentare con vari grafici

    4. Rilevanza delle caratteristiche -> identificata la variabile target vedere come questa varia nel dataset effettuando varie operazioni

        *Studiare la correlazione con la variabile target e le altre(definite variabili ESPLICATIVE o INDIPEDENTI)
