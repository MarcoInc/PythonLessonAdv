La riproducibilità computazione è una grande sfida specie nei lavori di scienza dei dati in cui tutti i dati dei dataset non è completamente accessibile

Gli articoli tecnici terminano con

    *approccio usato**

    *breve citazione sulla metodologia

    *approccio teorico

Sono quindi pochi dettagli che non permettono ad altri di riprodurre i lavoro

La macanza di riproducibilità computazionale quindi è dovuta anche alla mancanza di documentazione



Jupyter Notebook è molto utile per questo scopo

1. fornire un'introduzione dettagliata al problema;

    *buona introduzione

    *riepilogo

    *definizione del problema

2. documentare l'approccio e il flusso di lavoro;

    *definire il processo -> algoritmi usati

    *commentare il codice + manutenzione dei commenti per modifiche future

3. spiegare il canale dati;

    *da dove derivano i dati?

    *creare un dizionaro dei dati usati

4. descrivere le dipendenze;

    *dipedenze software
        *pacchetti e dipendenze usate (matplot,TensorFlow..)

    *dipedenze hardware 

5. utilizzare il controllo della versione del codice sorgente;

    *versioning del codice per poter guardare versioni precedenti del codice

6. rendere modulare il processo.

    *evitare la duplicazione del codice

    *usare delle librerie personali per richiamare il codice dove serve
